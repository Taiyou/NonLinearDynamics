# 9.1 Computing fractal dimensions

# capacity dimension
Capacity dimension is to determine how many a small box $\epsilon$ can cover the original set.
\begin{eqnarray*}
 d_{cap} = lim_{\epsilon -> 0^+} \frac{\log(N(\epsilon))}{\log(\frac{1}{\epsilon})}
\end{eqnarray*}

## Capacity dimension can be determined by the relationship in the middle of capacity dimension
\begin{eqnarray*}
    N(\epsilon) \propto \epsilon^{-d_{cap}}
\end{eqnarray*}

\begin{eqnarray}
x_{curr} \rightarrow ceiling\big[ \frac{x_{curr} - \min x}{\epsilon} \big]
\end{eqnarray}

Algorithm to count points in each box $\epsilon$. This is expensive to calculte.

# correlation dimension
First calculation

Approximating correlation dimension by correlation sum
\begin{eqnarray}
 C(\epsilon) \frac{1}{N(N-1)} \sum_{i=1}^N \sum_{j=i+1}^N \Theta (\epsilon - || x_i - x_j ||)
\end{eqnarray}

\begin{eqnarray}
    \Theta = 
    \begin{cases}
        1 & \text{if x $\geq$ 0 } \\
        0 & \text{if x < 0 } \\
    \end{cases}
\end{eqnarray}

\begin{eqnarray*}
    C(\epsilon) \propto \epsilon^{d_{corr}}
\end{eqnarray*}
