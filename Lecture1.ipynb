```python
R = [2,3.4,3.72];
n = 200;
```

```python
def PlotLogisticMap(R,n):
    rlength= len(R);
    results1 = np.zeros((rlength,n));
    results2 = np.zeros((rlength,n));
    
    # calculate the logistic map
    for r in range(rlength):
        # set initial values
        x1_new = 0.2; x2_new = 0.200001;
        results1[r,0] = x1_new;
        results2[r,0] = x2_new;
        for ii in range(1, n):
            x1_new = logistic(x1_new, R[r])
            x2_new = logistic(x2_new, R[r])
            results1[r,ii] = x1_new; results2[r,ii] = x2_new; 
    
    # plot the results of abs difference between two lines
    for r in range(rlength):
        plt.figure(r);
        plt.plot(np.abs(results1[r,:]-results2[r,:]));
        
    plt.show();
    return results1,results2
```

```python
[result1, result2] = PlotLogisticMap(R,n);
```

```python
np.mean(abs(result1[2,:]-result2[2,:]))
```

```python
n = 2000;
[result1, result2] = PlotLogisticMap(R,n);
np.mean(abs(result1[2,:]-result2[2,:]))
```
