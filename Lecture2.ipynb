The overall structure will look similar to that of Figure 1 below (although not quite identical; we'll get to that later in this quiz).

```python
StepSize = (4-2.4)/0.01;
# r = (2.4, 4)
R = np.linspace(2.4, 4,StepSize+1);
iteration = 1000;
threshold = 5;
```

```python
# logistic map
def logistic(x,r):
    return r*x*(1-x)
```

```python
# plots for bifurcation diagram of logistic map
def LogisticBifPlot(R, iteration, threshold):
    # parameter r
    for r in R:
        # initial values
        x_new = 0.2;
        results = [];
        for it in range(iteration):
            x_new = logistic(x_new, r);
            if it > threshold-1:
                results.append(x_new);
        plt.plot(r, np.array([results]), '.b');
        
    plt.show()
```

```python
LogisticBifPlot(R, iteration, threshold)
```

```python
iteration = 1000;
threshold = 500;
# only plot threshold = 500;
LogisticBifPlot(R, iteration,threshold)
```
